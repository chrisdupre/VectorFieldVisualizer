```python
#DO NOT MODIFY
import matplotlib.pyplot as plt
import numpy as np
```

We are gonna be graphing vector fields! 
As a quick introduction, here is a table for how to convert handwritten expressions into something the computer will understand. 
$$\begin{align}
x+y && \text{x+y}\\
x\times y && \text{x*y}\\
x(y+z) && x*(y+z)\\
x^y && \text{x**y}\\
\sin(x), \cos(y) && \text{np.sin(x)},\text{np.cos(y)}\\
e^{x}, \log(x) && \text{np.exp(x)}, \text{np.log(x)}
\end{align}$$


Lets have you import your vector field below! Please **only modify the cell directly below this one until you see text like this again** We will adopt the convention that 
$$x' = v,\ \  y' = w $$
x_min, x_max, y_min, y_max simply modify what values of $x$ and $y$  are plotted in the vector plot. Feel free to modify them to include any equilibrium points you would like to see. fine controls the fineness of the grid. The larger the value, the more points are plotted and the longer it will take to run (although you may get better pictures). fine =100 will normally be fine


After you have inputted your expressions, press Shift enter twice to see your plot

```python
x_min = -1
x_max = 1
y_min = -1
y_max = 1
def v(x,y):
    return x-y**3 ##Modify v here

def w(x,y):
    return -y ##Modify w here
fine = 100
```

```python
#DO NOT MODIFY
x = np.linspace(x_min,x_max,fine)
y = np.linspace(y_min,y_max,fine)
v_plot = np.array([v(x[i],y[j]) for j in range(fine) for i in range(fine)]).reshape(fine,fine)
w_plot = np.array([w(x[i],y[j]) for j in range(fine) for i in range(fine)]).reshape(fine,fine)
plt.streamplot(x,y,v_plot,w_plot)
```
