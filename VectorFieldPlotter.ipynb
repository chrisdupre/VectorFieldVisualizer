```python
#DO NOT MODIFY
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
x = sym.Symbol('x')
y = sym.Symbol('y')
v = sym.Function('v')(x,y)
w = sym.Function('w')(x,y)
```

```python
x_min = -1
x_max = 1
y_min = -1
y_max = 1
v = x**5 - y
w = sym.sin(x) + sym.exp(y)
fine = 100
```

```python
#DO NOT MODIFY
x_plot = np.linspace(x_min,x_max,fine)
y_plot = np.linspace(y_min,y_max,fine)
v_plot = np.array([float(v.evalf(subs = {x:x_plot[i],y:y_plot[j]})) for i in range(fine) for j in range(fine)]).reshape(fine,fine)
w_plot = np.array([float(w.evalf(subs = {x:x_plot[i],y:y_plot[j]})) for i in range(fine) for j in range(fine)]).reshape(fine,fine)
plt.streamplot(x_plot,y_plot,v_plot,w_plot)
```


# Calculating the linearization 
Okay! Now that we are able to see a plot for our vector field, let us turn to the linear analysis of its fixed points. 

Remember that for the vector field $x' = v(x,y), y' = w(x,y)$, we have the following form of its linearization
$$ J = \begin{bmatrix} \frac{dv}{dx} && \frac{dv}{dy} \\ \frac{dw}{dx} && \frac{dw}{dy} \end{bmatrix} $$
The next cell will run this for you automatically given the vector fields you gave above

```python
J = sym.Matrix([[sym.diff(v,x), sym.diff(v,y)],[sym.diff(w,x),sym.diff(w,y)]])
J
```

After obtaining the symbolic form, we need to actually evaluate the linearization at the equilibrium points ($v(x,y) = 0, w(x,y) = 0$) . Plug in one of the equilibrium points (with x_eval set to the x-coordinate and y_eval set to the y-coordinate) you found below to see the linearization at that point

```python
x_eval = 1
y_eval = 2
J.evalf(subs = {x:x_eval,y:y_eval})
```

The next cell will determine the eigenvalues of the Matrix at that point for you. Make sure you can also find the eigenvalues without this check. The output is formatted as follows. If the linearization has eigenvalues $\lambda_1$ and $\lambda_2$ with algebraic multiplicities $a_1, a_2$ respectively, then the output will be $$\{\lambda_1:a_1, \lambda_2:a_2\}$$

```python
J.evalf(subs={x:x_eval,y:y_eval}).eigenvals()
```
